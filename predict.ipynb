```python
# Splits our data into seasons, uses previous season to predict new season
# Start = 2, meaning we need at least 2 seasons
# We will use 2016 and 2017 to predict season 2018
def backtest(data, model, predictors, start=2, step=1):
    all_predictions = []
    seasons = sorted(data["season"].unique())
    for i in range(start, len(seasons), step):
        season = seasons[i]

        # The cross validation set for previous seasons
        train = data[data["season"] < season]
        # The set to test our data on
        test = data[data["season"] == season]

        # Fitting our model, takes training data and target 
        # The target is the value we want to predict
        model.fit(train[predictors], train["target"])

        # Generate predictions on test set
        # Convert numpy array to a pandas series
        preds = model.predict(test[predictors])
        preds = pd.Series(preds, index=test.index)

        # Create a dataframe to compare target and prediction
        combined = pd.concat([test["target"], preds], axis=1)
        combined.columns = ["actual", "prediction"]

        all_predictions.append(combined)
    return pd.concat(all_predictions)        
```

```python
predictions = backtest(df, rr, predictors)
```

```python
# Create a metric to tell us how accurate our model is
from sklearn.metrics import accuracy_score

predictions = predictions[predictions["actual"] != 2]
accuracy_score(predictions["actual"], predictions["prediction"])
```

```python
# Create a function to find team averages per team, performance for the last 10 games
# Finding rolling averages - basically we find the mean values/performance of the team in the last 10 games
# Grouping by team/season

def find_team_averages(team):
    # Select only numeric columns
    numeric_cols = team.select_dtypes(include='number')
    # Apply rolling mean to these columns
    rolling = numeric_cols.rolling(15).mean()
    return rolling

df_rolling = df_rolling.groupby(["team", "season"], group_keys=False).apply(find_team_averages)

    
```

```python
# Concatenating the rolling averages with the other metrics
rolling_cols = [f"{col}_15" for col in df_rolling.columns]
df_rolling.columns = rolling_cols

df = pd.concat([df, df_rolling], axis=1)
```

```python
# Letting the model know if the next game is home/away and who the opponent is to make better judgement
# Take the value of the next game and push it back 1 row
def shift_col(team, col_name):
    next_col = team[col_name].shift(-1)
    return next_col

def add_col(df, col_name):
    return df.groupby("team", group_keys=False).apply(lambda x: x[col_name].shift(-1))

# Creating new column which will tell us if its home next or away, next opp/date
df["home_next"] = add_col(df, "home")
df["team_opp_next"] = add_col(df, "team_opp")
df["date_next"] = add_col(df, "date")


                                                      
```
